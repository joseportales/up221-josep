```python
# import libraries
import geopandas as gpd
import pandas as pd
import plotly.express as px
```

```python
tax = pd.read_csv("data/Assessor_Chinatown.xlsx - Assessor_Chinatown.csv")
```

```python
tax.head()
```

```python
# convert data to geodata
tax = gpd.GeoDataFrame(tax, geometry=gpd.points_from_xy(tax.Longitude, tax.Latitude))
```

```python
tax["Imp Current Roll Year"].unique().tolist()
```

```python
tax.plot()
```

```python
fig = px.scatter_mapbox(tax,
                        lat='Latitude',
                        lon='Longitude',
                        # size='Total_Number_of_Seats',
                        color='Imp Current Value',
                        # size_max=20,
                        # center = {'lat':center_lat,'lon':center_lon},
                        mapbox_style="carto-positron")

# options on the layout
fig.update_layout(
    width = 800,
    height = 800,
    title = "Flexible office spaces",
    title_x=0.5 # aligns title to center
)

fig.show()
```
